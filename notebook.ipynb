```python
def create_dataset(X, time_steps=1):
    Xs, ys = [], []
    size = X.shape[0]
    for j in range(size):
        for i in range(188 - time_steps):
            #print(f"i = {i}")
            #print(f"j = {j}")
            v = X.iloc[j][i:(i + time_steps)].values
            #print(np.matrix(v).T)
            Xs.append(np.matrix(v).T)        
            ys.append(X.iloc[j][i + time_steps])
    return np.array(Xs), np.array(ys)
```
